```python
import pandas as pd
import numpy as np

#
data = {
    'Order Date': pd.date_range(start='2023-01-01', periods=100, freq='D'),
    'Category': np.random.choice(['Electronics', 'Furniture', 'Office Supplies'], 100),
    'Sales': np.random.randint(100, 1000, 100),
    'Profit': np.random.randint(10, 200, 100),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], 100)
}

df = pd.DataFrame(data)

```

```python
df.to_csv('sales_data.csv', index=False)
```

```python
print("Data created and saved as sales_data.csv!")
df.head()
```

```python
#%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px

df = pd.read_csv('sales_data.csv')
st.title("Business Dashboard")
fig = px.bar(df, x='Category', y='Sales', title="Sales by Category")
st.plotly_chart(fig)

#st.show()
```
